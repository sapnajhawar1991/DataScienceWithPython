# Assignment # 1: Sentiment Analysis for Obama and Trump Tweets 

## OBAMA Tweets

Step 1: In the first step, I would like to read Positive_word.txt file and make a list of positive_words.

```python
# assuming the location of file containin all the positive words to be as below. The file has each positive word on a new line
h =  open('C:\Users\sapkit\Documents\Data Science with Python\Homework1\Positive_word.txt', 'r')
```

```python
positive_words = []                 # declare the list of positive_words
for positive in h: 
    positive = positive.strip() 
    positive_words.append(positive)
print positive_words
```

Step 2: In the next step,reading the negative_word.txt and make a list of negative_words.

```python
# assuming the list of all the negative words to be in a file as below. The file has each negative word on a new line
q=open('C:\\Users\\sapkit\\Documents\\Data Science with Python\\Homework1\\negative_words.txt', 'r')
```

```python
negative_words = []   # declaring list that will contain all negative words
for negative in q:
    negative = negative.strip()
    negative_words.append(negative)
print negative_words
```

Step 3: Reading the obama.txt file and clean it to evaluate Sentiment Analysis

```python
# reading obama tweets from the file. 
```

```python
f = open('C:\Users\sapkit\Documents\Data Science with Python\Homework1\obama.txt','r') 
```

```python
# making a list of tweetwords by filtering the file of obama tweets using the list of positive and negative words
```

```python
tweetwords = [] #list
for line in f: 
    list_of_words_in_line = line.split()
    for word in list_of_words_in_line:
        word = word.strip(':;,.?"!@()#*').lower()
        # Pick a word only if its either positive or negative. Also making an assumption and removing word trump from the list
        # as trump is both used as a name and also a positive word.
        if (word in positive_words or
            word in negative_words) and word != 'trump':
            tweetwords.append(word)
            
print tweetwords
```

Step 4: Creating a Dictionary out of the list of tweetwords where key is the words and value will be the number of times(frequency) the words will be repeated. 

```python
my_dict = {}        # declaring the dictionary by the name my_dict
# code below will create a dictionary with key as the word and value as the number of times that word occurs in the tweets.
for i in range(len(tweetwords)):
    my_dict[tweetwords[i]] = tweetwords.count(tweetwords[i])
print (my_dict)
```

Step 5: Sorting the Dictionary by value in descending order(arranging it in a most frequent words said in the tweet) and giving it a variable name: sortByValueDict

```python
# We will sort the dictionary on values to form a list with top values first. Then we can pick the top 10 from this list.
sortByValueDict = sorted(my_dict.items(), key = lambda t: -t[1])
print sortByValueDict
```

Step 6: Picking up the top ten words out of sortByValueDict

```python
# We pick only the top 10 words from the list above
Top_10_words = sortByValueDict[:10]
Top_10_words # list of tuple
```

Step 7: Assinging +1 for positive sentiment word and -1 for negative sentiment word and evaluating the final sentiment value.

```python
# Assigning a sentiment of +1 for each positive word and -1 for each negative word in the top 10 words list.
sentiment = 0 
for words in Top_10_words:
    if words[0] in positive_words:
        print 'assigning +1 for positive sentiment %s' % words[0]
        sentiment += 1
    if words[0] in negative_words:
        print 'assigning -1 for negative sentiment %s' % words[0]
        sentiment -= 1
print 'The final sentiment value is %d' %sentiment
```
